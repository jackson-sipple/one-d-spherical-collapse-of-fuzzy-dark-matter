```python
import numpy as np
from scipy.integrate import trapz
```

```python
def L_c(M, z, L0, M1, p, q, r):
    return L0 * (M/M1)**p/(1+(M/M1)**q) * ((1+z)/7)**r  

def phi_L_given_M(L, M, z, sigma, L0, M1, p, q, r):
        prefactor = 1/(np.sqrt(2*np.pi)*sigma*L)
        L_c_val = L_c(M,z,L0,M1,p,q,r)
        exponent = -np.log(L/L_c_val)**2 / (2*sigma**2)
        return prefactor * np.exp(exponent)

def phi_L(L, z, dn_dM, sigma, L0, M1, p, q, r):
        lo, hi = 1e7, 1e15
        integrand = lambda M: dn_dM(M, z) * phi_L_given_M(L, M, z, sigma, L0, M1, p, q, r)
        return trapz_integrate(integrand, lo, hi, logspace=True)

def trapz_integrate(integrand, lo, hi, n_intervals=100, logspace=False):
    integrand_fn = integrand
    if logspace:
        lo, hi = np.log(lo), np.log(hi)
        integrand_fn = lambda logx: np.exp(logx) * integrand(np.exp(logx))
    x = np.linspace(lo, hi, n_intervals)
    y = integrand_fn(x)
    dx = x[1] - x[0]
    return trapz(y=y, x=x, dx=dx)
```
